### -Filling new values for null in features from LS-RB

```python
df[df["Position"]=='GK'].count()
```

```python
empty = df[df['LS'].isna()]
empty[empty['Position']!='GK'].count()
```

```python
# so among 2085 missing values, 2025 of them are goal keepers and the rest are players with unknown position
# but before we filling, we need to transfer dtype from object to float 
miss_1 = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW',
       'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM',
       'CDM', 'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']
df_no_miss1 = df[~ df['LS'].isna()]
expertise = df_no_miss1['LS']
df['Expertise']=expertise.str.slice(start = -1).astype(int)
for feature in miss_1:
     df[feature] = df_no_miss1[feature].str.slice(start = 0,stop=2).astype(int)
```

```python
# Now we have features from LS-RB all as float number and we are going to fill with average value
for feature in miss_1:
    value = np.round(df[feature].mean(),0)
    df[feature].fillna(value, inplace = True)
```

### -Filling new values for null in features from Crossing -GKReflexes

```python
miss_2 = ['Crossing',
       'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling',
       'Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration',
       'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower',
       'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
       'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
       'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
       'GKKicking', 'GKPositioning', 'GKReflexes']
```

```python
# we just check Crossing to see if this is actually a group that have missing values in all miss_2 and it does seem right
df[df['Crossing'].isnull()]
```

```python
for feature in miss_2:
    value = np.round(df[feature].mean(),0)
    df[feature].fillna(value, inplace = True)
df.info()
```

### -Filling new values for null in features Release Clause 

```python
# First we need to transfer all release clause to integer
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        return float(x.replace('K', '')) * 1000
    if 'M' in x:
        return float(x.replace('M', '')) * 1000000
    return float(x)

df['Release Clause'] = df['Release Clause'].str.slice(start = 1).apply(value_to_float)
```

```python
# now we fill in with average release clause value
value = np.round(df['Release Clause'].mean(),0)
df['Release Clause'].fillna(value, inplace = True)
```
