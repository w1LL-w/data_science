### -Player's average overall based on Nationality
- To ensure the data make sense and we would only compare countries with more that 200 registered players.
- 200 is just an option

```python
# so we will first filter out all countries that have less than 200 players
df_1 = df.groupby(['Nationality'],as_index = False).count()
nation_200 = df_1[df_1['ID'] > 199]
nation_200
```

```python
# With nation_200, we make sure only country with more than 200 players shown its overall average
df_mean = df.groupby(['Nationality'], as_index = False).mean()
nation_200_avg = df_mean[df_mean['Nationality'].isin(nation_200['Nationality'])]
nation_200_avg_sorted = nation_200_avg.sort_values(by = ['Overall'])


# plot
sns.barplot(x = 'Overall', y= 'Nationality', data = nation_200_avg_sorted )
```

```python
nation_200_avg
```

#### idea: as we have all average value of player in the country with more than 200 palyers, hece we could visualize all player's traits

```python
nation_200_avg.columns
```

```python
# Lets look at avg value of features
# define the features we want fisrt
attri =[ 'Age', 'Overall', 'Potential',
       'International Reputation', 'Weak Foot', 'Skill Moves', 'Jersey Number',
       'Crossing', 'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys',
       'Dribbling', 'Curve', 'FKAccuracy', 'LongPassing', 'BallControl',
       'Acceleration', 'SprintSpeed', 'Agility', 'Reactions', 'Balance',
       'ShotPower', 'Jumping', 'Stamina', 'Strength', 'LongShots',
       'Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties',
       'Composure', 'Marking', 'StandingTackle', 'SlidingTackle']

# fetures related to Goalkeepers were deleted as not appropriate, Id, Nationnality, Special are also removed
# p is a random palette
p = ["Blues_d","RdBu", "Set1","muted","ch:2.5,-.2,dark=.3"]
for feature in attri:
    plt.figure()
    sort = nation_200_avg.sort_values(by = [feature])
    sns.barplot(x = feature, y= 'Nationality', data = sort, palette=random.choice(p))
    plt.xlabel('Average '+ feature )
    plt.show()
    
    
```

#### How top10 players compare to non-top10 players?

```python
df.loc[df['Nationality'].isin(nation10),'top10']= 'yes'
df.loc[~df['Nationality'].isin(nation10),'top10']= 'no'
category10 = df.groupby(['top10']).mean()
category10
```

```python
for feature in attri:
    plt.figure()
    sns.barplot(x = category10.index, y= feature, data = category10)
    plt.show()
```

```python
df.groupby(['top10']).count()['ID']
```

#### Findings: 
- There are 9860 players from non-top-10 countries and 8347 from top-10 countries
- Players from top10 have a overall higher rating, around 2 more
- Each feature does not have a significant different, usually have a difference of 2
- Interesting is that players with higher rating tends to choose smaller jersey number
